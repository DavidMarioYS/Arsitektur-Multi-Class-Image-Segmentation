```python
# Rotasi 15, 30, 45, 60, 90
import os
from PIL import Image
import shutil

# Path folder input preprocessing ori dan ground truth
preprocessed_ori_data = "C:\\Users\\Admin\\Documents\\SKRIPSI\\DATASET\\PREPROCESSING\\256\\split\\original\\val\\original"
preprocessed_gt_data = "C:\\Users\\Admin\\Documents\\SKRIPSI\\DATASET\\PREPROCESSING\\256\\split\\original\\val\\ground_truth"

# Path folder untuk hasil augmentasi ori dan ground truth
final_aug_ori = 'C:\\Users\\Admin\\Documents\\SKRIPSI\\DATASET\\PREPROCESSING\\256\\split\\augmentation\\val\\original'
final_aug_gt = 'C:\\Users\\Admin\\Documents\\SKRIPSI\\DATASET\\PREPROCESSING\\256\\split\\augmentation\\val\\ground_truth'

# Path ke folder output untuk hasil augmentasi images dan ground truth
rotate_ori = 'C:\\Users\\Admin\\Documents\\SKRIPSI\\DATASET\\PREPROCESSING\\256\\split\\augmentation\\val\\rotate\\original'
rotate_gt = 'C:\\Users\\Admin\\Documents\\SKRIPSI\\DATASET\\PREPROCESSING\\256\\split\\augmentation\\val\\rotate\\ground_truth'

# Membuat folder tujuan jika belum ada
os.makedirs(final_aug_ori, exist_ok=True)
os.makedirs(final_aug_gt, exist_ok=True)
os.makedirs(rotate_ori, exist_ok=True)
os.makedirs(rotate_gt, exist_ok=True)

# Fungsi untuk menyalin file dari satu folder ke folder tujuan
def copy_files(source_folder, destination_folder):
    for filename in os.listdir(source_folder):
        # Mendapatkan path lengkap file asal
        source_path = os.path.join(source_folder, filename)

        # Memastikan bahwa hanya file yang disalin, bukan folder
        if os.path.isfile(source_path):
            # Mendapatkan path tujuan
            destination_path = os.path.join(destination_folder, filename)

            # Menyalin file ke folder tujuan
            shutil.copy2(source_path, destination_path)

# Menyalin file dari folder preprocessing ori dan gt ke folder final augmentasi
copy_files(preprocessed_ori_data, final_aug_ori)
copy_files(preprocessed_gt_data, final_aug_gt)

print(f"Semua file dari {preprocessed_ori_data} telah disalin ke {final_aug_ori}.")
print(f"Semua file dari {preprocessed_gt_data} telah disalin ke {final_aug_gt}.")

# Fungsi untuk melakukan rotasi pada gambar dan menyimpannya
def rotate_and_save(image, angle, original_name, extension, output_folder):
    rotated_image = image.rotate(angle)  # Rotate image
    new_filename = f"{original_name}_rotated_{angle}{extension}"  # New file name
    rotated_image.save(os.path.join(output_folder, new_filename))  # Save the rotated image

# Loop melalui semua file gambar di folder input
for filename in os.listdir(preprocessed_ori_data):
    if filename.endswith('.png'):  # Filter untuk file gambar
        # Dapatkan path file ori
        file_path_image = os.path.join(preprocessed_ori_data, filename)

        # Ganti prefix ORI_ dengan GT_ untuk ground truth
        gt_filename = filename.replace('ORI_', 'GT_')
        file_path_gt = os.path.join(preprocessed_gt_data, gt_filename)

        # Cek apakah file ground truth ada
        if not os.path.exists(file_path_gt):
            print(f"File ground truth tidak ditemukan: {file_path_gt}")
            continue  # Lewati jika file ground truth tidak ditemukan

        # Buka gambar ori dan ground truth
        img = Image.open(file_path_image)
        gt = Image.open(file_path_gt)

        # Pisahkan nama file dan ekstensi
        file_name, file_extension = os.path.splitext(filename)

        # Lakukan rotasi dengan sudut yang diinginkan
        for angle in [15, 30, 45, 60, 90]:
            # Rotasi dan simpan gambar ori
            rotate_and_save(img, angle, file_name, file_extension, rotate_ori)
            # Rotasi dan simpan ground truth
            rotate_and_save(gt, angle, file_name.replace('ORI_', 'GT_'), file_extension, rotate_gt)

# Menyalin file dari folder rotasi ke folder final augmentasi ori dan gt
copy_files(rotate_ori, final_aug_ori)
copy_files(rotate_gt, final_aug_gt)

print("Proses augmentasi rotasi dan salin file ke folder final augmentasi selesai.")
```

```python
import os
from PIL import Image
from PIL import ImageEnhance
import numpy as np
import shutil

# Path folder input preprocessing ori dan ground truth
preprocessed_ori_data = "C:\\Users\\Admin\\Documents\\SKRIPSI\\DATASET\\PREPROCESSING\\256\\split\\original\\val\\original"
preprocessed_gt_data = "C:\\Users\\Admin\\Documents\\SKRIPSI\\DATASET\\PREPROCESSING\\256\\split\\original\\val\\ground_truth"

# Folder untuk hasil augmentasi
brightness_ori = 'C:\\Users\\Admin\\Documents\\SKRIPSI\\DATASET\\PREPROCESSING\\256\\split\\augmentation\\val\\brightness\\original'
brightness_gt = 'C:\\Users\\Admin\\Documents\\SKRIPSI\\DATASET\\PREPROCESSING\\256\\split\\augmentation\\val\\brightness\\ground_truth'

hue_ori = 'C:\\Users\\Admin\\Documents\\SKRIPSI\\DATASET\\PREPROCESSING\\256\\split\\augmentation\\val\\hue\\original'
hue_gt = 'C:\\Users\\Admin\\Documents\\SKRIPSI\\DATASET\\PREPROCESSING\\256\\split\\augmentation\\val\\hue\\ground_truth'

saturation_ori = 'C:\\Users\\Admin\\Documents\\SKRIPSI\\DATASET\\PREPROCESSING\\256\\split\\augmentation\\val\\saturation\\original'
saturation_gt = 'C:\\Users\\Admin\\Documents\\SKRIPSI\\DATASET\\PREPROCESSING\\256\\split\\augmentation\\val\\saturation\\ground_truth'

# Path folder untuk hasil augmentasi ori dan ground truth
final_aug_ori = 'C:\\Users\\Admin\\Documents\\SKRIPSI\\DATASET\\PREPROCESSING\\256\\split\\augmentation\\val\\original'
final_aug_gt = 'C:\\Users\\Admin\\Documents\\SKRIPSI\\DATASET\\PREPROCESSING\\256\\split\\augmentation\\val\\ground_truth'

# Membuat folder tujuan jika belum ada
folders = [brightness_ori, brightness_gt, hue_ori, hue_gt, saturation_ori, saturation_gt, final_aug_ori, final_aug_gt]
for folder in folders:
    os.makedirs(folder, exist_ok=True)

# Function to change hue
def change_hue(image, hue_shift):
    img_hsv = image.convert('HSV')
    np_img_hsv = np.array(img_hsv, dtype=np.uint8)
    hue_channel = np_img_hsv[..., 0].astype(np.int16)
    hue_channel = (hue_channel + hue_shift) % 256
    np_img_hsv[..., 0] = hue_channel.astype(np.uint8)
    img_rgb = Image.fromarray(np_img_hsv, 'HSV').convert('RGB')
    return img_rgb

# Function to change saturation
def change_saturation(image, saturation_factor):
    enhancer = ImageEnhance.Color(image)
    return enhancer.enhance(saturation_factor)

# Function to change brightness
def change_brightness(image, brightness_factor):
    enhancer = ImageEnhance.Brightness(image)
    return enhancer.enhance(brightness_factor)

# Get all image files in the preprocessed original data folder
image_files = [f for f in os.listdir(preprocessed_ori_data) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

# Brightness, Hue, and Saturation factors
brightness_factors = [0.5, 1.0, 1.5]
hue_shifts = [10, -10]
saturation_factors = [0.5, 1.0, 1.5, 2.0]

# Loop to process each image
for image_file in image_files:
    image_path = os.path.join(preprocessed_ori_data, image_file)
    image_name, image_ext = os.path.splitext(image_file)

    # Change filename to find ground truth
    gt_image_file = f"GT_{image_name[4:]}.png"
    gt_image_path_png = os.path.join(preprocessed_gt_data, gt_image_file)

    # Check if the ground truth file exists in .png format
    if os.path.exists(gt_image_path_png):
        gt_image_path = gt_image_path_png
    else:
        print(f"Ground truth image for {image_name} not found.")
        continue

    # Brightness augmentation
    for factor in brightness_factors:
        augmented_image_brightness = change_brightness(Image.open(image_path), factor)
        brightness_name = f"{image_name}_brightness_{factor}.png"
        augmented_image_brightness.save(os.path.join(brightness_ori, brightness_name))

        # Copy ground truth image
        gt_augmented_name = f"GT_{image_name[4:]}_brightness_{factor}.png"
        shutil.copy(gt_image_path, os.path.join(brightness_gt, gt_augmented_name))

    # Hue augmentation
    for hue_shift in hue_shifts:
        augmented_image_hue = change_hue(Image.open(image_path), hue_shift)
        hue_shift_name = f"{image_name}_hue_shift_{hue_shift}.png"
        augmented_image_hue.save(os.path.join(hue_ori, hue_shift_name))

        # Copy ground truth image
        gt_augmented_name = f"GT_{image_name[4:]}_hue_shift_{hue_shift}.png"
        shutil.copy(gt_image_path, os.path.join(hue_gt, gt_augmented_name))

    # Saturation augmentation
    for factor in saturation_factors:
        augmented_image_saturation = change_saturation(Image.open(image_path), factor)
        saturation_name = f"{image_name}_saturation_{factor}.png"
        augmented_image_saturation.save(os.path.join(saturation_ori, saturation_name))

        # Copy ground truth image
        gt_augmented_name = f"GT_{image_name[4:]}_saturation_{factor}.png"
        shutil.copy(gt_image_path, os.path.join(saturation_gt, gt_augmented_name))

# Copy augmented images to final destination
for src_folder, dest_folder in [(brightness_ori, final_aug_ori), (hue_ori, final_aug_ori), (saturation_ori, final_aug_ori)]:
    for augmented_file in os.listdir(src_folder):
        if augmented_file.lower().endswith(('.png', '.jpg', '.jpeg')):
            shutil.copy(os.path.join(src_folder, augmented_file), os.path.join(dest_folder, augmented_file))

for src_folder, dest_folder in [(brightness_gt, final_aug_gt), (hue_gt, final_aug_gt), (saturation_gt, final_aug_gt)]:
    for gt_file in os.listdir(src_folder):
        if gt_file.lower().endswith(('.png', '.jpg', '.jpeg')):
            shutil.copy(os.path.join(src_folder, gt_file), os.path.join(dest_folder, gt_file))

print(f"Semua augmentasi telah selesai dan disalin ke folder final.")
```

```python
import os
from PIL import Image, ImageFilter
import numpy as np
import shutil

# Folder paths
preprocessed_ori_data = "C:\\Users\\Admin\\Documents\\SKRIPSI\\DATASET\\PREPROCESSING\\256\\split\\original\\val\\original"
preprocessed_gt_data = "C:\\Users\\Admin\\Documents\\SKRIPSI\\DATASET\\PREPROCESSING\\256\\split\\original\\val\\ground_truth"

# Path folder untuk Gaussian blur augmentation
blur_ori = 'C:\\Users\\Admin\\Documents\\SKRIPSI\\DATASET\\PREPROCESSING\\256\\split\\augmentation\\val\\blur\\original'
blur_gt = 'C:\\Users\\Admin\\Documents\\SKRIPSI\\DATASET\\PREPROCESSING\\256\\split\\augmentation\\val\\blur\\ground_truth'

# Path folder untuk hasil augmentasi ori dan ground truth
final_aug_ori = 'C:\\Users\\Admin\\Documents\\SKRIPSI\\DATASET\\PREPROCESSING\\256\\split\\augmentation\\val\\original'
final_aug_gt = 'C:\\Users\\Admin\\Documents\\SKRIPSI\\DATASET\\PREPROCESSING\\256\\split\\augmentation\\val\\ground_truth'

# Create blur_ori and blur_gt folders if they don't exist
os.makedirs(blur_ori, exist_ok=True)
os.makedirs(blur_gt, exist_ok=True)

# Function to apply Gaussian blur
def apply_gaussian_blur(image, radius):
    return image.filter(ImageFilter.GaussianBlur(radius))

# Get all image files in the preprocessed original data folder
image_files = [f for f in os.listdir(preprocessed_ori_data) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

# Gaussian blur radii
blur_radii = [1, 2, 3]  # Adjust radii as needed

# Loop to process each image
for image_file in image_files:
    image_path = os.path.join(preprocessed_ori_data, image_file)
    image_name, image_ext = os.path.splitext(image_file)  # Split filename and extension

    # Change filename to find ground truth
    gt_image_file = f"GT_{image_name[4:]}.png"  # Replace 'ORI_' with 'GT_' and use .png format
    gt_image_path_png = os.path.join(preprocessed_gt_data, gt_image_file)

    # Check if the ground truth file exists in .png format
    if os.path.exists(gt_image_path_png):
        gt_image_path = gt_image_path_png
    else:
        print(f"Ground truth image for {image_name} not found.")
        continue  # Skip if ground truth image is not found

    # Loop for Gaussian blur augmentation
    for radius in blur_radii:
        augmented_image_blur = apply_gaussian_blur(Image.open(image_path), radius)
        blur_name = f"{image_name}_blur_{radius}.png"  # Output always .png
        augmented_image_blur.save(os.path.join(blur_ori, blur_name))

        # Copy ground truth image with the corresponding name, with "_mask" format
        gt_augmented_name = f"GT_{image_name[4:]}_blur_{radius}.png"  # Output mask always .png
        shutil.copy(gt_image_path, os.path.join(blur_gt, gt_augmented_name))

# Create final destination folders if they don't exist
os.makedirs(final_aug_ori, exist_ok=True)
os.makedirs(final_aug_gt, exist_ok=True)

# Copy augmented images to final destination
for augmented_file in os.listdir(blur_ori):
    if augmented_file.lower().endswith(('.png', '.jpg', '.jpeg')):
        shutil.copy(os.path.join(blur_ori, augmented_file), os.path.join(final_aug_ori, augmented_file))

# Copy augmented ground truth images to final destination
for gt_file in os.listdir(blur_gt):
    if gt_file.lower().endswith(('.png', '.jpg', '.jpeg')):
        shutil.copy(os.path.join(blur_gt, gt_file), os.path.join(final_aug_gt, gt_file))

print(f"File dari {blur_ori} telah disalin ke {final_aug_ori}.")
print(f"File dari {blur_gt} telah disalin ke {final_aug_gt}.")
```

```python
# import os
from PIL import Image
from collections import Counter

# Define the folder paths
final_aug_ori = 'C:\\Users\\Admin\\Documents\\SKRIPSI\\DATASET\\PREPROCESSING\\256\\split\\augmentation\\val\\original'
final_aug_gt = 'C:\\Users\\Admin\\Documents\\SKRIPSI\\DATASET\\PREPROCESSING\\256\\split\\augmentation\\val\\ground_truth'

def analyze_images(folder_path):
    total_images = 0
    total_size = [0, 0]  # Width, Height
    total_pixels = 0
    color_modes = []
    file_formats = []

    for filename in os.listdir(folder_path):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
            total_images += 1
            img_path = os.path.join(folder_path, filename)
            with Image.open(img_path) as img:
                width, height = img.size
                total_size[0] += width
                total_size[1] += height
                total_pixels += width * height
                color_modes.append(img.mode)
                file_formats.append(img.format)

    # Calculate average size
    if total_images > 0:
        avg_size = (total_size[0] // total_images, total_size[1] // total_images)
        avg_total_pixels = total_pixels / total_images
        most_common_color_mode = Counter(color_modes).most_common(1)[0][0]
        most_common_format = Counter(file_formats).most_common(1)[0][0]
    else:
        avg_size = (0, 0)
        avg_total_pixels = 0
        most_common_color_mode = None
        most_common_format = None

    return {
        "Total Gambar": total_images,
        "Rata-rata Ukuran Gambar": avg_size,
        "Rata-rata Total Piksel": avg_total_pixels,
        "Mode Warna Paling Sering": most_common_color_mode,
        "Format File": most_common_format
    }

# Analyze both folders
results = {
    "Original": analyze_images(final_aug_ori),
    "Ground Truth": analyze_images(final_aug_gt)
}

# Display results
for folder_name, info in results.items():
    print(f"Hasil Analisis untuk Folder {folder_name}:")
    print(f"Total Gambar: {info['Total Gambar']}")
    print(f"Rata-rata Ukuran Gambar: ({info['Rata-rata Ukuran Gambar'][0]:.2f} x {info['Rata-rata Ukuran Gambar'][1]:.2f})")
    print(f"Rata-rata Total Piksel: {info['Rata-rata Total Piksel']:.2f}")
    print(f"Mode Warna Paling Sering: {info['Mode Warna Paling Sering']}")
    print(f"Format File: {info['Format File']}\n")

# Ensure that the number of images in the Original and Ground Truth folders is the same
assert results["Original"]["Total Gambar"] == results["Ground Truth"]["Total Gambar"], "Jumlah gambar di folder ORI dan GT tidak sama"

# Ensure that the average image size in both folders is the same
assert results["Original"]["Rata-rata Ukuran Gambar"] == results["Ground Truth"]["Rata-rata Ukuran Gambar"], "Rata-rata ukuran gambar di folder ORI dan GT tidak sama"

# Ensure that the average total pixels in both folders are the same
assert results["Original"]["Rata-rata Total Piksel"] == results["Ground Truth"]["Rata-rata Total Piksel"], "Rata-rata total piksel di folder ORI dan GT tidak sama"

# Ensure that the most common color mode in both folders is the same
assert results["Original"]["Mode Warna Paling Sering"] == results["Ground Truth"]["Mode Warna Paling Sering"], "Mode warna paling sering di folder ORI dan GT tidak sama"

# Ensure that the file formats in both folders are the same
assert results["Original"]["Format File"] == results["Ground Truth"]["Format File"], "Format file di folder ORI dan GT tidak sama"

print("\nSemua pengujian berhasil! Folder original dan ground_truth memiliki data gambar yang sama.")
```
